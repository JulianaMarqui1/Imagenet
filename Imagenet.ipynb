#Imagenet

Modelos utilizados: MobileNetV2 e ResNet50

##Etapa 01

Importar as bibliotecas

```python
import tensorflow
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

##Etapa 02

Criar as redes MobileNetV2 e ResNet50

```python
base_model_MobileNet = MobileNetV2(weights='imagenet')
base_model_ResNet    = ResNet50(weights='imagenet')
```

##Etapa 03

Testar com uma imagem aleatória

```python
img_path = '/content/Persa.webp'  # colocar o nome da figura 1

img_MobileNet = image.load_img(img_path, target_size=(224, 224))
x_MobileNet   = image.img_to_array(img_MobileNet)

img_ResNet = image.load_img(img_path, target_size=(224, 224))
x_ResNet   = image.img_to_array(img_ResNet)

x_MobileNet = np.expand_dims(x_MobileNet, axis=0)
img_preprocessed_MobileNet = tensorflow.keras.applications.mobilenet_v2.preprocess_input(x_MobileNet)
print(x_MobileNet.shape)

x_ResNet = np.expand_dims(x_ResNet, axis=0)
img_preprocessed_ResNet = tensorflow.keras.applications.resnet50.preprocess_input(x_ResNet)

features_MobileNet = base_model_MobileNet.predict(img_preprocessed_MobileNet)
features_ResNet    = base_model_ResNet.predict(img_preprocessed_ResNet)
```

```python
img_path2 = '/content/sphynx.webp'  # colocar o nome da figura 2

img_MobileNet2 = image.load_img(img_path2, target_size=(224, 224))
x_MobileNet2   = image.img_to_array(img_MobileNet2)

img_ResNet2 = image.load_img(img_path2, target_size=(224, 224))
x_ResNet2   = image.img_to_array(img_ResNet2)

x_MobileNet2 = np.expand_dims(x_MobileNet2, axis=0)
img_preprocessed_MobileNet2 = tensorflow.keras.applications.mobilenet_v2.preprocess_input(x_MobileNet2)
print(x_MobileNet2.shape)

x_ResNet2 = np.expand_dims(x_ResNet2, axis=0)
img_preprocessed_ResNet2 = tensorflow.keras.applications.resnet50.preprocess_input(x_ResNet2)

features_MobileNet2 = base_model_MobileNet.predict(img_preprocessed_MobileNet2)
features_ResNet2    = base_model_ResNet.predict(img_preprocessed_ResNet2)
```

```python
img_path3 = '/content/Maine-Coon.webp'  # colocar o nome da figura 2

img_MobileNet3 = image.load_img(img_path3, target_size=(224, 224))
x_MobileNet3   = image.img_to_array(img_MobileNet3)

img_ResNet3 = image.load_img(img_path3, target_size=(224, 224))
x_ResNet3   = image.img_to_array(img_ResNet3)

x_MobileNet3 = np.expand_dims(x_MobileNet3, axis=0)
img_preprocessed_MobileNet3 = tensorflow.keras.applications.mobilenet_v2.preprocess_input(x_MobileNet3)
print(x_MobileNet3.shape)

x_ResNet3 = np.expand_dims(x_ResNet3, axis=0)
img_preprocessed_ResNet3 = tensorflow.keras.applications.resnet50.preprocess_input(x_ResNet3)

features_MobileNet3 = base_model_MobileNet.predict(img_preprocessed_MobileNet3)
features_ResNet3    = base_model_ResNet.predict(img_preprocessed_ResNet3)
```

# MobileNetV2


```python
#primeira imagem
resultado_MobileNet = tensorflow.keras.applications.mobilenet_v2.decode_predictions(features_MobileNet, top=5)[0]

resultado_top5_MobileNet = np.array([resultado_MobileNet[0][2], resultado_MobileNet[1][2],
                                     resultado_MobileNet[2][2], resultado_MobileNet[3][2],
                                     resultado_MobileNet[4][2]])
labels_top5_MobileNet = np.array([resultado_MobileNet[0][1], resultado_MobileNet[1][1],
                                   resultado_MobileNet[2][1], resultado_MobileNet[3][1],
                                   resultado_MobileNet[4][1]])

print('Resultado:\nLabels: ', labels_top5_MobileNet, '\nProbabilidades:', resultado_top5_MobileNet)

plt.figure(figsize=(15, 3))
plt.subplot(1, 2, 1)
plt.imshow(img_MobileNet)
plt.subplot(1, 2, 2)
plt.bar(range(5), resultado_top5_MobileNet, tick_label=labels_top5_MobileNet)
plt.show()

#segunda imagem
resultado_MobileNet2 = tensorflow.keras.applications.mobilenet_v2.decode_predictions(features_MobileNet2, top=5)[0]

resultado_top5_MobileNet2 = np.array([resultado_MobileNet2[0][2], resultado_MobileNet2[1][2],
                                     resultado_MobileNet2[2][2], resultado_MobileNet2[3][2],
                                     resultado_MobileNet2[4][2]])
labels_top5_MobileNet2 = np.array([resultado_MobileNet2[0][1], resultado_MobileNet2[1][1],
                                   resultado_MobileNet2[2][1], resultado_MobileNet2[3][1],
                                   resultado_MobileNet2[4][1]])

print('Resultado:\nLabels: ', labels_top5_MobileNet2, '\nProbabilidades:', resultado_top5_MobileNet2)

plt.figure(figsize=(15, 3))
plt.subplot(1, 2, 1)
plt.imshow(img_MobileNet2)
plt.subplot(1, 2, 2)
plt.bar(range(5), resultado_top5_MobileNet2, tick_label=labels_top5_MobileNet2)
plt.show()

#terceira imagem
resultado_MobileNet3 = tensorflow.keras.applications.mobilenet_v2.decode_predictions(features_MobileNet3, top=5)[0]

resultado_top5_MobileNet3 = np.array([resultado_MobileNet3[0][2], resultado_MobileNet3[1][2],
                                     resultado_MobileNet3[2][2], resultado_MobileNet3[3][2],
                                     resultado_MobileNet3[4][2]])
labels_top5_MobileNet3 = np.array([resultado_MobileNet3[0][1], resultado_MobileNet3[1][1],
                                   resultado_MobileNet3[2][1], resultado_MobileNet3[3][1],
                                   resultado_MobileNet3[4][1]])

print('Resultado:\nLabels: ', labels_top5_MobileNet3, '\nProbabilidades:', resultado_top5_MobileNet3)

plt.figure(figsize=(15, 3))
plt.subplot(1, 2, 1)
plt.imshow(img_MobileNet3)
plt.subplot(1, 2, 2)
plt.bar(range(5), resultado_top5_MobileNet3, tick_label=labels_top5_MobileNet3)
plt.show()
```

Acertou as duas primeiras imagens. Na terceira não identificou a raça mas identificou uma característica do gato correta. Pode ser a forma como foi categorizado no dataset.
